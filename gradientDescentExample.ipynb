# Contour Plot of the function $f(x)=x_{1}x_{2}e^{-x_{1}^{2}-x_{2}^{2}}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def myFun(x1, x2):
    return x1 * x2 * np.exp(-x1**2 - x2**2)

# Generate a grid of values
x1 = np.linspace(-2, 2, 100)
x2 = np.linspace(-2, 2, 100)
x1mesh, x2mesh = np.meshgrid(x1, x2)
fmesh = myFun(x1mesh, x2mesh)

# Create the contour plot with 20 colored contour lines
plt.figure()
contour = plt.contour(x1mesh, x2mesh, fmesh, 20, cmap='jet')

# Add contour labels
plt.clabel(contour, inline=True, fontsize=8)

# Add labels and a title
plt.title(r'Contour lines of $f(x_1, x_2) = x_1 x_2 e^{-x_1^2 - x_2^2}$')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

# Display the plot
plt.show()
```

```python
def myFun(x):
    return np.exp(-x[0]*x[0]-x[1]*x[1])*(x[0]*x[1])


fig = plt.figure()
ax = fig.gca(projection="3d")

xmesh, ymesh = np.mgrid[-2:2:50j,-2:2:50j]
fmesh = myFun(np.array([xmesh, ymesh]))
# Add a title to the plot
plt.title(r'3D Plot of the function $f(x_1, x_2) = x_1 x_2 e^{-x_1^2 - x_2^2}$')
ax.plot_surface(xmesh, ymesh, fmesh,cmap='jet')
```
